#### 24) Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe)

```python
df24_1=pd.DataFrame({'playerID':df['playerID'],'yearID':df['yearID'],'Age':(2018-df['birthYear'])})
df24_2=pd.merge(df,df24_1,how='left',on=['playerID','yearID']).reset_index()   
```

```python
df24_3=df24_2[(df24_2['HR']>=50)].groupby(['playerID','nameFirst','nameLast','Age','yearID']).agg({'HR':'sum'}).reset_index().sort_values('Age')
df24_4=df24_3[(df24_3['Age']==df24_3['Age'].min())]
df24_4
```
